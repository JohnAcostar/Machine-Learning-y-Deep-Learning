2. Corra el modelo aumentando las epocas hasta que llegue a converger. Conteste: ¿Cuánto es el accuracy?

```python
#ANTES DE CORRER ESTA CELDA, CORRER EL CODIGO ANTERIOR PARA ESTABLECER LAS FUNCIONES Y VARIABLES
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = TransformerEncoderClassifier(len(vocab)).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

train_losses, val_accs = [], []

for epoch in range(35): #ELEGI UN NUMERO ARBITRARIO PARA OBSERVAR CUANDO LLEGSA LA CONVERGENCIA DEL LOSS
    model.train()
    total_loss = 0
    for x, y in tqdm(train_dl, desc=f"Epoch {epoch+1}"):
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        preds = model(x)
        loss = criterion(preds, y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    train_losses.append(total_loss / len(train_dl))

    # Evaluación
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x, y in val_dl:
            x, y = x.to(device), y.to(device)
            preds = model(x)
            correct += (preds.argmax(1) == y).sum().item()
            total += y.size(0)
    acc = correct / total
    val_accs.append(acc)
    print(f"Epoch {epoch+1}: loss={train_losses[-1]:.4f}, val_acc={acc:.4f}")

# =====================================
# Gráficas
# =====================================
plt.figure(figsize=(8,4))
plt.plot(train_losses, label="Train Loss")
plt.plot(val_accs, label="Validation Accuracy")
plt.legend()
plt.xlabel("Época")
plt.title("Entrenamiento Transformer SST-2")
plt.show()
```

El accuracy fue de 0.7569 al final de la convergencia.

3. Explique en que parte del código se hace el padding, y si este aplica solo al entrenar o también al momento de usar el modelo.

De lo observado en el codigo, el padding se produce dentro de la función encode_sentence,

if len(ids) < max_len:
    
    ids += [vocab["<pad>"]] * (max_len - len(ids))

Este es el momento en el que la frase de padding con el token <pad> sucede, y se detiene hasta alcanzar max_len (default = 50). Después de tokenizar y convertir los tokens en índices de vocabulario, se comprueba si la secuencia es más corta que la longitud fija. Si lo es, el código rellena/padding las posiciones restantes con el índice <pad>. Ahora bien, El padding se aplica antes de que los datos entren en DataLoader, en la clase class SST2Dataset(Dataset) dentro de getitem. Dado que, el conjunto de datos devuelve tensores ya con padding, el dataLoader procesa por lotes/stacks y los datos de entrenamiento están con el padding. Por consiguiente, en la sección de inferencia de escribir palabras

x = encode_sentence(phrase, vocab, tokenizer).unsqueeze(0).to(device)

Este vuelve a llamar a encode_sentence, por lo que el padding también se produce tambien al predecir una sola frase.


4. Pruebe 10 frases en ingles distintas (5 positivas y 5 negativas). Reporte que tal se comporta el modelo.



```python
model.eval()
while True:
    phrase = input("\nEscribe una frase (o 'salir'): ")
    if phrase.lower() == 'salir':
        break
    x = encode_sentence(phrase, vocab, tokenizer).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(x)
        pred = output.argmax(1).item()
    print("Positiva" if pred == 1 else "Negativa")
```

El modelo clasifica muy bien las frases con sentimiento directo y preciso, por ejemplo “I really enjoyed this movie” (positiva) o “I hated the ending so much” (negativa). Esto significa y sugiere  que aprendió palabras fuertemente asociadas a cada clase durante el entrenamiento. Sin embargo, falla en oraciones más mixtas o con emociones neutrales y complejas, Para dar un ejemplo, “was a interested the story” es gramaticalmente incorrecta y la marca como negativa, y “It was much better than I expected”, como positiva.

5. Investigue sobre BERT y señale las diferencias que tiene con el codigo utilizado en este laboratorio.

Principalmente, BERT se le aplica un preentrenamiento masivo con miles de millones de tokens utilizando el modelado de lenguaje enmascarado y la Next Sentence Prediction, lo que le permite adquirir una mayor capacidad lingüística antes de ser ajustado para una tarea posterior. Por el contrario, el modelo del laboratorio se entrena desde cero únicamente con SST-2, por lo que debe aprender los patrones lingüísticos directamente a partir de un pequeño conjunto de datos. Igualmente, desde el punto de vista arquitectónico, BERT es mucho más complejo, en capacidad computacional, BERT-base tiene 12 capas de codificador, 12 cabezas de atención y un tamaño oculto de 768 dimensiones, mientras que el modelo del laboratorio contiene un bloque de codificador con dos cabezas de atención y un estado oculto de 64 dimensiones, lo que lo hace más pequeño en los parametros de entrada. Adicionalmente, la tokenización y el embedding también difieren, BERT utiliza un tokenizador de subpalabras WordPiece que maneja las palabras complejas e incluye positional embedding de hasta 512 caracteres, mientras que el modelo de laboratorio utiliza un tokenizador básico de espacios en blanco con un vocabulario simple de 10 000 palabras y una longitud máxima de 50. Además, BERT depende de tokens especiales como [CLS], para la clasificación, [SEP], para separar segmentos, y [MASK], para el preentrenamiento,, mientras que el modelo de laboratorio solo utiliza <pad> y <unk> y extrae el significado de las frases mediante el agrupamiento medio del embedding de tokens. Por ultimo, BERT también incorpora metodos de atención para evitar prestar atención a los tokens de relleno y presenta una regularización AdamW, mientras que el modelo de laboratorio no utiliza el enmascaramiento de atención.
