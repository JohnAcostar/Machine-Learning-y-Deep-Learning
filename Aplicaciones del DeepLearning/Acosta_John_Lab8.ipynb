En primer lugar, señalo que el modelo lineal puede captar tendencias a gran escala y generalizar mejor el conjunto de pruebas que los modelos RNN/GRU/LSTM. Es decir, una RNN/GRU/LSTM puede alcanzar un R² de entrenamiento muy alto al hacer un overfitting, o sobre ajuste, de los datos y las fluctuaciones de las ventanas de muestra. No obstante, esa volatilidad e imprediciblidad de las acciones aprendidas por los modelos no se mantienen en el test [1]. Por ese motivo, los modelos recurrentes mejoraron el R² de entrenamiento, pero no el R² de prueba. En contraste, una regresión lineal puede alcanzar un R² elevado para la predicción del día siguiente, ya que los precios de las acciones están fuertemente autocorrelacionados, lo que demuestra que la secuencia temporal contiene suficiente información para realizar previsiones a corto plazo [2]. Igualmente, menciona que el valor de la recurrencia surge una vez que se elimina el componente absolutista y de modelar la tendencia como una regresión lineal, y se le pide a la red que aprenda una estructura secuencial de mayor frecuencia. Por último, menciona que las RNN, las GRU y las LSTM no son simplemente máquinas de memoria, sino que aprenden la representación de secuencias en la recurrencia. Finalmente, las LSTM separan explícitamente la memoria a largo y corto plazo a través del estado de la celda [3]. Esto, permite a los modelos de deep learning adaptarse a datos temporales complejos y los mecanismos de atención.

```python
import itertools
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import time

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class LSTMClf(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return torch.sigmoid(out)

def train_classification_torch(model, X_train_np, y_train_np, X_test_np, y_test_np, epochs=200, batch_size=64, lr=1e-3):
    model = model.to(device)
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
    criterion = nn.BCELoss()
    Xtr = torch.tensor(X_train_np, dtype=torch.float32)
    ytr = torch.tensor(y_train_np.reshape(-1,1), dtype=torch.float32)
    Xte = torch.tensor(X_test_np, dtype=torch.float32)
    yte = torch.tensor(y_test_np.reshape(-1,1), dtype=torch.float32)
    train_ds = TensorDataset(Xtr, ytr)
    train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    best_state, best_test_acc = None, -np.inf
    for epoch in range(epochs):
        model.train()
        for xb, yb in train_dl:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
        model.eval()
        with torch.no_grad():
            preds = model(Xte.to(device)).cpu().numpy().ravel()
            preds_bin = (preds > 0.5).astype(int)
            test_acc = accuracy_score(y_test_np, preds_bin)
        if test_acc > best_test_acc:
            best_test_acc = test_acc
            best_state = model.state_dict()
    model.load_state_dict(best_state)
    return best_test_acc

hidden_sizes = [32, 64, 128, 256]
num_layers = [1, 2, 3, 4]
acc_matrix = np.zeros((len(hidden_sizes), len(num_layers)))

for i, h in enumerate(hidden_sizes):
    for j, l in enumerate(num_layers):
        model = LSTMClf(hidden_size=h, num_layers=l)
        acc = train_classification_torch(model, Xc_train_rnn, yc_train, Xc_test_rnn, yc_test, epochs=200, lr=5e-3)
        acc_matrix[i, j] = acc
        print(f"hidden_size={h}, num_layers={l}, acc={acc:.4f}")

plt.figure(figsize=(8,6))
sns.heatmap(acc_matrix, annot=True, fmt=".3f", xticklabels=num_layers, yticklabels=hidden_sizes, cmap="YlOrRd")
plt.xlabel("num_layers", color="firebrick")
plt.ylabel("hidden_size", color="firebrick")
plt.title("LSTM Heatmap del accuracy", color="firebrick")
plt.show()
```
