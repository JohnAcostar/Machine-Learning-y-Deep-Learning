2. Explique que diferencia existe entre *np.ravel()* y *np.flatten()*, y mencione su uso dentro del c√≥digo y el porque es importante.

Como se menciona y se explica en la documentaci√≥n de numpy, np.ravel() regresa una matriz unidimensional contigua del mismo subtipo que a, con forma (a.size,) (Numpy.Ravel ‚Äî NumPy v2.2 Manual, 2019.). Asimismo se se√±ala que la funcion es equivalente a usar al .reshape(-1). Para dar un ejemplo el siguiente 

```python
import numpy as np
z = np.array([[1, 50, 100], [150, 200, 250]])
print(np.ravel(z))
print(z.reshape(-1))
y = np.ravel(z)
y[0] =10
print(y)
print(z)
```

Ahora bien, la diferencia con np.flatten() se encuentra que flatten siempre devuelve una nueva copia de la matriz en 1D, sin afectar al objeto original si este nueva variable es modificada. En contraste, ravel() no hace una copia del objeto cuando una nueva variable es declarada con la funcion. En consecuencia, las operaciones posteriores que afecten una variable generaran que cambien el objeto original.

3. Busque informaci√≥n sobre retropropagaci√≥n y como esto ayuda a volver ajustables los pesos y sesgos de las capas ocultas.

Inicialmente, la retropropagaci√≥n tiene el proposito de busca ajustar los pesos de una red neuronal para minimizar la funci√≥n de p√©rdida, este proceso implica calcular los gradientes de p√©rdida con respecto a cada peso (Damadi et¬†al., 2023). Esto se debe a que, el poder de una red neuronal esta ligado a las capas ocultas, si la red no tiene las suficiente capas ocultas para extraer las caracteristicas o comportamiento de los datos, poseera, en consecuencia, un poder de representacion o prediccion de los datos limitada. No obstante, si queremos aumentar el numero de capas ocultas aumentara tambien el numero de parametros de la red, el modelo poseera mas pesos W y bias B, tendra mas complejidad algoritmica y m√°s costo computacional (Sharkawy, 2024). Ahora bien, a causa de que no es sugerible optimizar cada parametro de la red manualmente, se emplea la retropropagaci√≥n para hallar los parametros que mejor se desempe√±an o predigan el modelo con los datos de entrada. 

En este sentido, es necesario comprender el descenso del gradiente, este metodo busca minimizar el error de una funci√≥n al calcular el error en un punto especifico y las derivadas parciales en ese punto. De este modo se obtiene el calculo de la direccion y la pendiente de la funci√≥n e iterativamente ir en direccion opuesta para hallar el minimo. Para nuestra red neuronal, es hallar el gradiente del error de cada peso en la red, y para un problema de Regresion lineal se tiene 2 parametros 
$\frac{\partial{error}}{\partial{W}}$ y $\frac{\partial{error}}{\partial{b}}$. Igualmente, el error o loss que se esta calculando, es una composici√≥n de varias funciones $L(f(z))$, $L$ es el loss o funcion de perdida, que en este caso es la entropia cruzada, $f()$ es la funcion de activacion que es sigmoide y $z$ es la salida de la neurona. Posteriormente se calcula para cada capa de neurona el nuevo peso (W) y el nuevo sesgo (b) ajustandose por el learning rate y la funcion de perdida. De esta forma, cuando se ajustan todos los parametros se completa una epoca, se vuelve a hacer un feedforward se calcula la retropropagaci√≥n y el descenso del gradiente de nuevo y se completa otra epoca en el entraniemto de los parametros.

4. Corre el modelo con distintos n√∫meros de capas densas *(1, 5, 10, 15, 20, 25, 30)*, y haga una gr√°fica que en el eje x tenga el n√∫mero de capas densas y en el eje y tenga el valor del *accuracy*. ¬øQue puede concluir al respecto?


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Leer el archivo Excel
df = pd.read_excel('datosDengueTangamandapio.xlsx')

# Extraer las columnas como vectores (arrays de NumPy)
x_real = df['Lluvias'].values.reshape(-1,1)
y_real = df['Pico de casos'].values.reshape(-1,1)

# Escalar entrada
X_escalado = (x_real - np.mean(x_real)) / np.std(x_real)
Y_escalado = y_real
X_escalado = np.ravel(X_escalado)
Y_escalado = np.ravel(Y_escalado)

# Funciones base
def hidden_layer(X, W1, b1):
    z = np.outer(X, W1.flatten()) + b1  # (n datos,N_lay)
    return 1 / (1 + np.exp(-z))         # sigmoide (n datos,N_lay)

def predict_hidd(X, w, b, W1, b1):
    h = hidden_layer(X, W1, b1)      
    y = h @ w + b            
    return 1 / (1 + np.exp(-y))

def loss_ce(y_pred, y_true):
    epsilon = 1e-8
    return -np.mean(y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))

def grad_ce(X, y_true, y_pred, W1, b1):
    h = hidden_layer(X, W1, b1)                      
    error = (y_pred - y_true)[:, np.newaxis]  
    dw = np.mean(h * error, axis=0)           
    db = np.mean(y_pred - y_true)             
    return dw, db

def train(X, y, loss_fn, grad_fn, N_lay, lr=0.1, epochs=1000):
    W1 = np.random.randn(1, N_lay)
    b1 = np.random.randn(N_lay)
    w = np.random.randn(N_lay)
    b = np.random.randn()
    for epoch in range(epochs):
        y_pred = predict_hidd(X, w, b, W1, b1)
        dw, db = grad_fn(X, y, y_pred, W1, b1)
        w -= lr * dw
        b -= lr * db
    return w, b, W1, b1

def accuracy_score(y_true, y_probs, umbral=0.5):
    y_pred = (y_probs >= umbral).astype(int)
    return np.mean(y_pred == y_true)

def evaluar_manual(y_true, y_probs, umbral=0.5,epsilon = 1e-8):
    y_pred = (y_probs >= umbral).astype(int)

    TP = np.sum((y_pred == 1) & (y_true == 1)) #Verdaderos positivos
    TN = np.sum((y_pred == 0) & (y_true == 0)) #Verdaderos negativos
    FP = np.sum((y_pred == 1) & (y_true == 0)) #Falsos positivos
    FN = np.sum((y_pred == 0) & (y_true == 1)) #Falsos negativos

    accuracy = (TP + TN) / len(y_true)
    precision = TP / (TP + FP + epsilon)
    recall = TP / (TP + FN + epsilon)
    f1 = 2 * precision * recall / (precision + recall + epsilon)

    print("üìä Evaluaci√≥n manual:")
    print(f"Accuracy  : {accuracy:.4f}")
    print(f"Precision : {precision:.4f}")
    print(f"Recall    : {recall:.4f}")
    print(f"F1-score  : {f1:.4f}")
    print("Matriz de confusi√≥n:")
    print(f"TP: {TP} | FP: {FP}")
    print(f"FN: {FN} | TN: {TN}")

# Experimentos con diferentes capas densas
layers_list = [1, 5, 10, 15, 20, 25, 30]
accuracies = []

for N_lay in layers_list:
    w, b, W1, b1 = train(X_escalado, Y_escalado, loss_ce, grad_ce, N_lay, lr=0.1, epochs=1000)
    probs = predict_hidd(X_escalado, w, b, W1, b1)
    acc = accuracy_score(Y_escalado, probs)
    accuracies.append(acc)
    # Obtener probabilidades
    probs_ce  = predict_hidd(X_escalado, w,  b,W1,b1)
    # Evaluaci√≥n
    print("\nEvaluaci√≥n modelo con Entrop√≠a Cruzada:")
    evaluar_manual(Y_escalado, probs_ce)

# Graficar resultado
plt.figure(figsize=(8,6))
plt.plot(layers_list, accuracies, marker='o', linestyle='--')
plt.xlabel("N√∫mero de capas densas (unidades ocultas)")
plt.ylabel("Accuracy")
plt.title("Precisi√≥n vs N√∫mero de capas densas")
plt.grid(True)
plt.show()

accuracies
```
